```python
import pandas as pd

df = pd.read_csv("Absenteeism-data.csv")
df.drop(columns=["ID"], inplace=True)
df.head()
```

```python
dummies = pd.get_dummies(df["Reason for Absence"], prefix="Grupo", drop_first=False)
df["Group_1"] = dummies.loc[:, "Grupo_1":"Grupo_14"].max(axis=1).astype(int)
df["Group_2"] = dummies.loc[:, "Grupo_15":"Grupo_17"].max(axis=1).astype(int)
df["Group_3"] = dummies.loc[:, "Grupo_18":"Grupo_21"].max(axis=1).astype(int)
df["Group_4"] = dummies.loc[:, "Grupo_22":"Grupo_28"].max(axis=1).astype(int)
df.drop(columns=["Reason for Absence"], inplace=True)
df.head()


```

```python
# Convertir a datetime y extraer el mes y dia
df["Fecha"] = pd.to_datetime(df["Date"], dayfirst=True)
df["Mes"] = df["Fecha"].dt.month
# Consideramos el Lunes como 1 hasta el Domingo = 7
df["DiaDeSemana"] = df["Fecha"].dt.weekday + 1
#eliminar date
df.drop(columns=["Date"], inplace=True)

#convertir a binario education
df["Education"] = df["Education"].apply(lambda x: 1 if x == 0 else 0)
df.head()
```

```python
# Guardamos los cambios en un nuevo
df.to_csv("df-cleaned.csv", index=False)
```
